# Step 2: Generate Dummy Radar and Precipitation Data

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Function to generate dummy 3D radar data and precipitation values
def generate_dummy_data(samples, radar_shape):
    radar_data = np.random.rand(samples, *radar_shape) * 100  # Radar data in range 0-100
    precipitation_data = np.random.rand(samples) * 10  # Precipitation in range 0-10 mm
    return radar_data, precipitation_data

# Configuration 1: Single radar value -> Precipitation
radar_data_1, precipitation_data_1 = generate_dummy_data(1000, (1,))

# Configuration 2: 3x3 radar grid -> Precipitation
radar_data_2, precipitation_data_2 = generate_dummy_data(1000, (3, 3))

# Configuration 3: 3x3x10 radar grid (10 elevation levels) -> Precipitation
radar_data_3, precipitation_data_3 = generate_dummy_data(1000, (3, 3, 10))

print(f"Shape of radar data for Config 1: {radar_data_1.shape}")
print(f"Shape of radar data for Config 2: {radar_data_2.shape}")
print(f"Shape of radar data for Config 3: {radar_data_3.shape}")
```

```python
# Visualizing the radar data for sanity checks
def visualize_radar_sample(radar_data, title):
    plt.imshow(radar_data.squeeze(), cmap='Blues')
    plt.title(title)
    plt.colorbar(label='Radar Reflectivity')
    plt.show()

# Sample visualizations
visualize_radar_sample(radar_data_2[0], "3x3 Radar Grid (Config 2)")
visualize_radar_sample(radar_data_3[0][:,:,0], "3x3 Radar Grid at Elevation 1 (Config 3)")
```

# Step 3: Define the Models

## Model 1: Single Radar Value to Precipitation Value (FCNN)

```python
import torch.nn as nn

class FCNN(nn.Module):
    def __init__(self):
        super(FCNN, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(1, 16),
            nn.ReLU(),
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.fc(x)

# Instantiate the model
model_1 = FCNN()
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, kernel_size=2)
        self.fc1 = nn.Linear(8 * 2 * 2, 1)  # Fully connected layer

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = x.view(x.size(0), -1)  # Flatten the feature map
        x = self.fc1(x)
        return x

# Instantiate the model
model_2 = CNN()
```

```python
class CNN3D(nn.Module):
    def __init__(self):
        super(CNN3D, self).__init__()
        self.conv3d = nn.Conv3d(1, 16, kernel_size=(2, 2, 2))
        
        # We will calculate the flattened size dynamically after the conv3d layer
        self._to_linear = None  # Placeholder to calculate flattened size
        self.fc1 = None  # Will be defined after getting flattened size

    def forward(self, x):
        x = torch.relu(self.conv3d(x))
        
        # Flatten the feature map after convolution
        if self._to_linear is None:
            self._to_linear = x.view(x.size(0), -1).size(1)  # Calculate the flattened size dynamically
            self.fc1 = nn.Linear(self._to_linear, 1)  # Initialize the fully connected layer based on the flattened size

        x = x.view(x.size(0), -1)  # Flatten the tensor
        x = self.fc1(x)
        return x

# Instantiate the model
model_3 = CNN3D()
```

# Step 4: Prepare Data for Training

```python
from torch.utils.data import DataLoader, TensorDataset

# Convert data to PyTorch tensors
radar_data_1_torch = torch.tensor(radar_data_1, dtype=torch.float32)
precipitation_data_1_torch = torch.tensor(precipitation_data_1, dtype=torch.float32).unsqueeze(1)

radar_data_2_torch = torch.tensor(radar_data_2, dtype=torch.float32).unsqueeze(1)  # Adding channel dimension
precipitation_data_2_torch = torch.tensor(precipitation_data_2, dtype=torch.float32).unsqueeze(1)

radar_data_3_torch = torch.tensor(radar_data_3, dtype=torch.float32).unsqueeze(1)  # Adding channel dimension
precipitation_data_3_torch = torch.tensor(precipitation_data_3, dtype=torch.float32).unsqueeze(1)

# Create DataLoader for each configuration
batch_size = 16
dataloader_1 = DataLoader(TensorDataset(radar_data_1_torch, precipitation_data_1_torch), batch_size=batch_size, shuffle=True)
dataloader_2 = DataLoader(TensorDataset(radar_data_2_torch, precipitation_data_2_torch), batch_size=batch_size, shuffle=True)
dataloader_3 = DataLoader(TensorDataset(radar_data_3_torch, precipitation_data_3_torch), batch_size=batch_size, shuffle=True)
```

# Step 5: Train the Models

```python
import torch
from tqdm import tqdm

def train_model(model, dataloader, num_epochs=5, model_name="Model"):
    print(f"Training {model_name}...")
    criterion = nn.MSELoss()  # Loss function
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        
        # Display progress bar for each epoch
        progress_bar = tqdm(dataloader, desc=f"Epoch [{epoch+1}/{num_epochs}] - {model_name}", leave=False)

        for radar_batch, precip_batch in progress_bar:
            optimizer.zero_grad()
            outputs = model(radar_batch)
            loss = criterion(outputs, precip_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        avg_loss = running_loss / len(dataloader)
        print(f"Epoch [{epoch+1}/{num_epochs}] - {model_name}, Loss: {avg_loss:.4f}")

# Train each model with model names
train_model(model_1, dataloader_1, model_name="FCNN (1 Radar Value)")
train_model(model_2, dataloader_2, model_name="CNN (3x3 Radar Grid)")
train_model(model_3, dataloader_3, model_name="3D CNN (3x3x10 Radar Grid)")
```

# Step 6: Visualize and Check results

```python
def visualize_results(model, dataloader):
    model.eval()
    with torch.no_grad():
        radar_batch, precip_batch = next(iter(dataloader))
        predictions = model(radar_batch)
        plt.plot(predictions.numpy(), label='Predictions')
        plt.plot(precip_batch.numpy(), label='Ground Truth')
        plt.legend()
        plt.show()

# Visualize results for each model
visualize_results(model_1, dataloader_1)
visualize_results(model_2, dataloader_2)
visualize_results(model_3, dataloader_3)
```
